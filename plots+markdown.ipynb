```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
high_lvl_df = pd.read_csv("/kaggle/input/isic-2024-challenge/train-metadata.csv", low_memory=False)
malignant_high_lvl_df = high_lvl_df[high_lvl_df['target'] == 1]
benign_high_lvl_df = high_lvl_df[high_lvl_df['target'] == 0]
```

We're loading the metadata dataset in order to perform exploratory data analysis on the metadata records and find any correlating features that can be used for our training algorithm. We are going to use data visualization in order to assist us in achieving results of our exploratory data analysis.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='age_approx', x='target', data=high_lvl_df, palette="Set2")

# Add title and labels
plt.title('Distribution of Age for Malignant and Benign Cases')
plt.xlabel('Label')
plt.ylabel('Age')
# Show the plot
plt.show()
```

First, we' relooking to find any correlation between age and malignancy of a given lesion. Based on the boxplot we graphed, there doesn't seem to be much defining differences for malignancy of a lesion given a person's age.

Next, looking at the graph for gender, it might look like lesions on male are likelier to be malignant. However, the entire sample size has the same proportion of male and female patients so we cannot conclude that gender has a correlation with malignancy of a lesion.

The average color contrast between the lesion and the skin from both the average and each individual color space from the LAB color space does not show any significant correlation with the malignancy. Moreover, different skin colors might alter the contrast between the lesion and the skin. Due to the importance of the color of the lesion to diagnose malignancy, the model will use the raw LAB color space of the lesion as features. The expectation is that the model would be able to detect correlation between each LAB color space that could increase or decrease the risk of malignancy.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='tbp_lv_nevi_confidence', x='target', data=high_lvl_df, palette="Set2")

# Add title and labels
plt.title('Distribution of Nevus Confidence Score for Malignant and Benign Cases')
plt.xlabel('Label')
plt.ylabel('Nevus Confidence Score')
# Show the plot
plt.show()
```

There's a significant difference between the distribution of nevus confidence score in malignant samples compared to the benign samples. Nevus confidence score is a metric to determine if a lesion is nevus (mole) or not where nevus are usually benign (Betz-Stablein et al., 2021).

Border asymmetry and irregularity doesn't seem to correlate well with the malignancy of a lesion. However, border irregularity is an important parameter for malignancy determination (Jaworek-Korjakowska, 2015).

In summary, we conclude that features that help with detecting the malignancy of a lesion are features that describe the location of the lesion, size of the lesion, color of the lesion, and the nevus confidence score associated to the particular lesion image. On the other hand, color contrast and border irregularity does not seem to correlate, but due to the importance of border irregularity for diagnosis, it will still be included as features in the model. Color contrast was not included as different skin color will alter the contrast. Moreover, the model will include the raw LAB color data of the lesion as features as the color of the lesion itself is more important of an indicator than the contrast.
